```python
import os
import pickle
import numpy as np

# 데이터 폴더 생성
os.makedirs('data/my_dataset', exist_ok=True)

# 신데렐라 텍스트
text_content = """옛날 옛적, 어느 나라에 아름답고 마음씨 착한 소녀가 살고 있었습니다. 소녀는 어릴 적 어머니를 잃었지만, 어머니가 남긴 사랑과 가르침을 마음속 깊이 간직하며 자랐습니다. 어머니는 살아생전 늘 이렇게 말하곤 했습니다. “어떤 어려움이 닥쳐도 친절함과 용기를 잃지 말아라.” 소녀는 그 말을 평생의 보물처럼 여겼습니다. 시간이 흐른 뒤 아버지는 새로운 아내를 맞이했습니다. 새어머니에게는 두 딸이 있었는데, 둘 다 외모를 꾸미는 데만 관심이 많고 성격은 매우 이기적이었습니다. 처음에는 모두가 함께 잘 지내는 듯했지만, 얼마 지나지 않아 새어머니와 두 딸은 소녀를 못마땅하게 여기기 시작했습니다. 그러던 어느 날, 소녀의 아버지마저 세상을 떠나고 말았습니다. 그날 이후 소녀의 삶은 완전히 달라졌습니다. 새어머니는 집안의 모든 일을 소녀에게 떠넘겼고, 두 의붓언니 역시 그녀를 하녀처럼 부렸습니다. 소녀는 새벽부터 밤늦게까지 청소를 하고, 빨래를 하고, 음식을 만들고, 장작을 나르며 하루를 보냈습니다. 특히 그녀는 늘 부엌 난로 곁에서 일해야 했기 때문에 옷과 얼굴에 재가 묻기 일쑤였습니다. 그래서 사람들은 그녀를 ‘재투성이 소녀’라는 뜻의 신데렐라라고 부르게 되었습니다. 비록 힘든 생활이 계속되었지만 신데렐라는 누구도 미워하지 않았습니다. 일이 끝난 밤이면 집 뒤뜰에 있는 어머니의 무덤을 찾아가 꽃을 놓고 조용히 이야기를 나누곤 했습니다. “어머니, 오늘도 열심히 살았어요.” 그때마다 신데렐라는 마음의 위안을 얻었습니다. 어느 날, 왕궁에서 큰 무도회가 열린다는 소식이 전국에 퍼졌습니다. 왕자는 결혼할 상대를 찾기 위해 나라의 젊은 여성들을 모두 초대했습니다. 마을 사람들은 흥분했고, 의붓언니들도 무도회 이야기에 들떠 있었습니다. “이번 기회에 왕자의 마음을 사로잡을 거야!” 언니들은 값비싼 옷을 고르고 장신구를 준비하며 하루 종일 거울만 들여다보았습니다. 신데렐라도 무도회에 가고 싶었습니다. 하지만 새어머니는 차갑게 말했습니다. “너 같은 아이가 무도회에 간다고? 집안일이나 해라.” 신데렐라는 실망했지만 포기하지 않았습니다. “저도 열심히 준비하면 갈 수 있을까요?” 그러자 새어머니는 렌틸콩 한 바구니를 재 속에 쏟아부으며 말했습니다. “이걸 두 시간 안에 모두 골라내면 생각해 보마.” 신데렐라는 묵묵히 콩을 골랐습니다. 그때 창문으로 새들이 날아와 그녀를 도와주었습니다. 덕분에 금세 일을 끝낼 수 있었지만 새어머니는 또 다른 핑계를 대며 결국 무도회에 보내주지 않았습니다. 저녁이 되자 새어머니와 의붓언니들은 화려하게 차려입고 마차를 타고 왕궁으로 떠났습니다. 홀로 남은 신데렐라는 눈물을 흘렸습니다. “나도 한 번쯤 아름다운 옷을 입고 춤을 춰 보고 싶었는데….” 바로 그 순간, 눈부신 빛이 부엌 안을 가득 채웠습니다. 신데렐라가 놀라 고개를 들자 한 요정이 나타났습니다. “왜 울고 있느냐, 착한 아이야?” 신데렐라는 지금까지의 일을 모두 이야기했습니다. 요정은 미소를 지으며 지팡이를 들어 올렸습니다. “걱정하지 마라. 오늘 밤 너도 무도회에 갈 수 있단다.” 요정은 커다란 호박을 황금빛 마차로 바꾸었습니다. 쥐들은 멋진 말이 되었고, 도마뱀들은 마부로 변했습니다. 그리고 낡은 옷을 향해 지팡이를 흔들자 눈부신 드레스가 나타났습니다. 마지막으로 반짝이는 유리구두 한 켤레가 신데렐라의 발에 신겨졌습니다. 신데렐라는 자신의 모습을 보고 깜짝 놀랐습니다. “정말 아름다워요!” 하지만 요정은 진지한 목소리로 말했습니다. “명심하렴. 마법은 자정이 되면 풀린다. 그전에 반드시 집으로 돌아와야 한다.” 신데렐라는 고개를 끄덕이고 왕궁으로 향했습니다. 무도회장은 수많은 촛불과 음악으로 가득했습니다. 신데렐라가 들어서자 모든 사람이 그녀를 바라보았습니다. 누구도 그녀가 누구인지 알지 못했지만, 너무나 우아하고 아름다웠기 때문입니다. 왕자 역시 그녀를 보자마자 시선을 떼지 못했습니다. 왕자는 곧장 다가와 정중히 인사했습니다. “이 춤을 저와 함께 추시겠습니까?” 신데렐라는 수줍게 손을 내밀었습니다. 두 사람은 음악에 맞춰 춤을 추었고, 이야기를 나누며 즐거운 시간을 보냈습니다. 왕자는 그녀의 아름다움뿐 아니라 따뜻하고 친절한 마음에 깊이 매료되었습니다. 시간은 순식간에 흘러갔습니다. 그때 멀리서 시계 종소리가 울리기 시작했습니다. 땡. 땡. 땡. 자정이었습니다. 신데렐라는 깜짝 놀라 계단을 향해 달려갔습니다. “잠깐만요! 이름이라도 알려 주세요!” 왕자가 외쳤지만 그녀는 멈추지 않았습니다. 급히 내려가던 중 유리구두 한 짝이 계단에 벗겨졌지만, 신데렐라는 그대로 마차에 올라탔습니다. 그리고 집에 도착하자마자 마법은 모두 풀려 버렸습니다. 다음 날 왕자는 유리구두를 들고 온 나라를 찾아다니게 했습니다. “이 구두가 맞는 여인이 바로 내가 찾는 사람이다.” 수많은 귀족 아가씨들이 구두를 신어 보았지만 맞지 않았습니다. 마내 왕자의 신하들이 신데렐라의 집에 도착했습니다. 의붓언니들은 어떻게든 구두를 신으려 했지만 발이 맞지 않았습니다. 그러자 신데렐라가 조심스럽게 말했습니다. “저도 신어 볼 수 있을까요?” 새어머니는 비웃었지만 왕자의 신하는 허락했습니다. 신데렐라가 구두를 신자 놀랍게도 그것은 마치 원래 그녀를 위해 만들어진 것처럼 완벽하게 맞았습니다. 그리고 신데렐라는 앞치마 주머니에서 다른 한 짝의 유리구두를 꺼내 보였습니다. 그 순간 모두는 그녀가 무도회의 그 여인이라는 사실을 알게 되었습니다. 왕자는 곧바로 달려와 기쁜 목소리로 말했습니다. “드디어 찾았습니다.” 얼마 후 두 사람은 성대한 결혼식을 올렸습니다. 신데렐라는 왕비가 되었지만 과거의 어려움을 잊지 않았습니다. 그녀는 가난한 사람들을 돕고, 누구에게나 친절하게 대했습니다. 사람들은 그녀를 아름다운 외모보다 따뜻한 마음으로 더 사랑했습니다. 그리고 신데렐라와 왕자는 서로를 아끼며 오래도록 행복하게 살았다고 합니다."""

# 파일 저장
with open('data/my_dataset/input.txt', 'w', encoding='utf-8') as f:
    f.write(text_content)

# 메타데이터 및 바이너리 파일 저장
chars = sorted(list(set(text_content)))
meta = {'vocab_size': len(chars), 'stoi': {ch:i for i,ch in enumerate(chars)}, 'itos': {i:ch for i,ch in enumerate(chars)}}
with open('data/my_dataset/meta.pkl', 'wb') as f:
    pickle.dump(meta, f)

data_ids = [meta['stoi'][ch] for ch in text_content]
n = len(data_ids)
np.array(data_ids[:int(n*0.9)], dtype=np.uint16).tofile('data/my_dataset/train.bin')
np.array(data_ids[int(n*0.9):], dtype=np.uint16).tofile('data/my_dataset/val.bin')
print("데이터 준비 완료! 다음 학습 코드를 실행하세요.")
```
